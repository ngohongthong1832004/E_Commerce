```python
import requests

def get_data(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # This will raise an error for HTTP errors (e.g., 404, 500)

        # Attempt to parse the JSON response
        try:
            data_dict = response.json()
        except ValueError as e:
            print(f"Failed to parse JSON response: {e}")
            return []  # Return an empty list if JSON parsing fails

        # Extract the "data" field if it exists, otherwise return an empty list
        data = data_dict.get("data", [])
        
        return data

    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
        return []
    except requests.exceptions.RequestException as req_err:
        print(f"Request error occurred: {req_err}")
        return []
```

```python
# write to file csv
def write_to_csv(data):
    df = pd.DataFrame(data)
    df.to_csv('products.csv', index=False)
    
# write to file json

def write_to_json(data):
    with open('data.json', 'w') as f:
        json.dump(data, f)
        
def save_to_parquet(data, filename="output.parquet"):
    # Convert data to a DataFrame
    df = pd.DataFrame(data)
    print(df)
    # Save DataFrame to Parquet
    df.to_parquet(filename, index=False)
    print(f"Data saved to {filename}")
```

```python
for i in range(len(df)):
    total_data = []
    
    # Check for 'link' column in DataFrame
    try:
        if 'link' not in df.columns:
            raise KeyError("The 'link' column is not present in the DataFrame.")
    except KeyError as e:
        print(f"Error: {e}")
        break  # Stop execution if 'link' column is missing
    
    try:
        param = df['link'].iloc[i].split("/c")[-1]
    except Exception as e:
        print(f"Error extracting category parameter: {e}")
        continue  # Move to the next iteration if there's an error
    
    for page in range(1, 50):
        try:
            url = f'https://tiki.vn/api/personalish/v1/blocks/listings?limit=300&category={param}&page={page}'
            data = get_data(url)
            if not data:
                print(f"No more data for category {param} on page {page}")
                break  # Stop paging if no data is returned
            total_data += data
        except Exception as e:
            print(f"Error processing page {page} of category {param}: {e}")
            continue  # Continue to the next page even if there's an error
    
    # Attempt to save data to Parquet
    try:
        save_to_parquet(total_data, f"category_{param}.parquet")
        print(f"Data saved to category_{param}.parquet")
    except Exception as e:
        print(f"Failed to save data to Parquet: {e}")
```
